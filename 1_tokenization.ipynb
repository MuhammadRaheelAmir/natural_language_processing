### Tokenization
- Data/text cleaning
- Breaking down or Interconversion of text
- Corpus, Documents, Words and Vocabulary

## Corpus
- Paragraph
- Collection of paragraphs
## Documents
- Sentences
- Collection of words
## Words
- Words 
- Smallest unit
## Vocabulary
- Unique words


```python
import nltk
```

```python
nltk.download('punkt_tab')
```

```python
corpus="""
I believe that learning should never stop, so I decided to start working on natural language processing. This field bridges human language with machine understanding, opening doors to exciting applications and research. By exploring NLP, I aim to strengthen my skills and stay connected with the evolving world of technology."""
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences=sent_tokenize(corpus)
print(sentences)
```

```python
from nltk.tokenize import word_tokenize
```

```python
words=word_tokenize(corpus)

print(words)
```

```python
from nltk.tokenize import wordpunct_tokenize
print(type(wordpunct_tokenize))
```

```python
words=wordpunct_tokenize(corpus)
print(words)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
words=tokenizer.tokenize(corpus)
print(words)
```
