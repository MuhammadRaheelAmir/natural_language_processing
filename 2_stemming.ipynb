### Stemming
- Text processing
- Reducing words to their root/base form
- Removing inflectional endings (e.g., running â†’ run)
- Removing suffixes and prefixes
- Helps normalize text for analysis

### Models
- Porter Stemmer
- Regexp Stemmer
- Snowball Stemmer


```python
from nltk.stem import PorterStemmer
```

```python
words=['eating','eats','eaten','eatable','write','writes','writing','written']
```

```python
portstemmer=PorterStemmer()
```

```python
for word in words:
    print(word+'---->'+portstemmer.stem(word))
```

```python
from nltk.stem import RegexpStemmer
```

```python
words=('eating','eats','eaten','eatable','write','writes','writing','written')
```

```python
regstemmer=RegexpStemmer('s$|es$|ies$|ing$|ed$|able$|')
```

```python
for word in words:
    print(word+'---->'+regstemmer.stem(word))
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+'---->'+snowstemmer.stem(word))
```
