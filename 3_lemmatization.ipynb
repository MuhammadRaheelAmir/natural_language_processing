### Lemmatization
- Normalization of words to their dictionary/base form (lemma)
- Considers morphology and context (part of speech)
- More accurate than stemming because it produces valid words
- Applied after tokenization and cleaning


### Parts of speech tagging - POS tag
- pos tag ----> Noun = n
- pos tag ----> Verb = v
- pos tag ----> Adverb = r
- pos tag ----> Adjective = a

```python
import nltk
from nltk.stem import wordnet
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('eating') 
```

```python
lemmatizer.lemmatize('eating',pos='v') # pos tag Verb

```

```python
words=['eat','eating','eats','eaten','eatable']
```

```python
for word in words:
    print(word+'---->'+lemmatizer.lemmatize(word,pos='v'))
```
